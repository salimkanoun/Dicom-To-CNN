```python
import SimpleITK as sitk 
import numpy as np 
import matplotlib.pyplot as plt 
from library_dicom.dicom_processor.tools.preprocessing import *
```

```python
image = '/media/deeplearning/Elements/classif_ct/nifti_gained/1.2.840.113619.2.55.3.279710477.870.1379571148.714_mip_ct_nifti_.nii'
```

```python
img = sitk.ReadImage(image)
array = sitk.GetArrayFromImage(img)
```

```python
array.shape
```

```python
f = plt.figure(figsize=(9,9))
axes = plt.gca()
axes.set_axis_off()
plt.imshow(array)
```

```python
np.max(array)
```

```python
# applied normalisation 
intensityWindowingFilter = sitk.IntensityWindowingImageFilter()
intensityWindowingFilter.SetOutputMaximum(1)
intensityWindowingFilter.SetOutputMinimum(0)
new_img = intensityWindowingFilter.Execute(img)
```

```python
new_array = sitk.GetArrayFromImage(new_img)
f = plt.figure(figsize=(9,9))
axes = plt.gca()
axes.set_axis_off()
plt.imshow(new_array, cmap='viridis')
```

```python
np.max(new_array)
```
